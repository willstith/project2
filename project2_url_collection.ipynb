I'm using a recursive function which re-calls itself every time the # of weight ratings is too low or if the game is an expansion, as well as every time it finishes adding a valid game while the number of games added is under 250.

```python
board_games_URLS = []
def url_compiler(target_size):
    
    """Compiles a list of URLs of random board games which are not expansions and which each have
        at least 10 user weight ratings.
        Input: INT, desired length of list
        Output: LIST of urls of length target_size"""
    
    randomGameURL = "https://boardgamegeek.com/boardgame/random"
    rejects = 0
    while len(board_games_URLS) < target_size:
        
        #go to BGG random game and parse its page
        page = requests.get(randomGameURL).text
        soup = BeautifulSoup(page,"lxml")
        
        #only include games rated by weight at least 10 times and exclude expansions 
        try:
            game_weight = int(re.search('(?<=\{"averageweight":).{1,50}(?=,"votes":)', str(soup)).group(0))
        except:
            try:
                game_weight = int(re.search('(?<=\{"averageweight":").{1,50}(?=,"votes":)', str(soup)).group(0))
            except:
                game_weight = 0
        try:
            weight_votes = int(re.search('(?<=,"votes":)[0-9]{1,9}(?=.*\}\})', str(soup)).group(0))
        except:
            try:
                weight_votes = int(re.search('(?<=,"votes":")[0-9]{1,9}(?=.*\}\})', str(soup)).group(0))
            except:
                weight_votes = 0
                
        expansion = re.search('Expansion for Base-game', str(soup))
        
        if (weight_votes < 10) or (expansion):
            for link in soup.find_all('link', rel='canonical'):
                #sanity check to know function isn't stuck
                print('FAIL: ', link.get("href"))
                rejects += 1
            time.sleep(5)
            url_compiler(target_size)
        
    
        #add the games w/ > 10 weight ratings to the list
        else:
            time.sleep(5)
            for link in soup.find_all('link', rel='canonical'):
                print('=====PASS: ', link.get("href"))
                board_games_URLS.append(link.get("href"))
            url_compiler(target_size)
                   
        return board_games_URLS
```
