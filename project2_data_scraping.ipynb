```python
from bs4 import BeautifulSoup
import requests
import time, os
import re
import pickle
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
with open('urls_686.pickle','rb') as read_file:
    BG_urls = pickle.load(read_file)
```

```python
len(BG_urls)
```

```python
data_dict = {}
```

```python
def BGG_scraper(game_urls_list):
    
    games_added = 0
    for game_url in game_urls_list:
        time.sleep(5)
    
        #go to game's API page
        gameid, url_name = game_url.split('/')[4], game_url.split('/')[5]
        game_API_url = f"https://boardgamegeek.com/xmlapi/boardgame/{gameid}?stats=1"
        ratings_href = f'/boardgame/{gameid}/{url_name}/ratings?rated=1'
        owners_href = f'/boardgame/{gameid}/{url_name}/ratings?status=own'
        plays_href = f'/playstats/thing/{gameid}'
        page = requests.get(game_API_url).text
    
        #parse XML
        root = ET.fromstring(page)
        bg = root.find('boardgame')
        stats = bg.find('statistics')
        ratings = stats.find('ratings')
    
        #grab data
        game_data = {}
        game_id = bg.get('objectid')
        for name in bg.findall('name'):
            if name.get('primary') == "true":
                game_data['name'] = name.text
            else:
                pass
        
        game_data['year'] = bg.find('yearpublished').text
        game_data['playingtime'] = bg.find('playingtime').text
        game_data['age'] = bg.find('age').text
        game_data['minplayers'] = bg.find('minplayers').text
        game_data['maxplayers'] = bg.find('maxplayers').text
        game_data['mintime'] = bg.find('minplaytime').text
        game_data['maxtime'] = bg.find('maxplaytime').text
    
        mechanics = []
        for mechanic in bg.findall('boardgamemechanic'):
            mechanics.append(mechanic.text)
        game_data['mechanics'] = mechanics
        game_data['num_mechanics'] = len(mechanics)
    
        categories = []
        for category in bg.findall('boardgamecategory'):
            categories.append(category.text)
        game_data['categories'] = categories
        game_data['num_categories'] = len(categories)
        
        subdomains = []
        for subdomain in bg.findall('boardgamesubdomain'):
            subdomains.append(subdomain.text)
        game_data['subdomains'] = subdomains
        game_data['num_subdomains'] = len(subdomains)
        
        game_data['weight'] = ratings.find('averageweight').text
        game_data['user_rating'] = ratings.find('average').text
        game_data['desc_wc'] = len(bg.find('description').text.split())
    
        game_data['kickstarted'] = 0
        for fam in bg.findall('boardgamefamily'):
            if fam.text == 'Crowdfunding: Kickstarter':
                game_data['kickstarted'] = 1
            else:
                pass
        
        #now use selenium to grab the number of forums for the game
        driver = webdriver.Chrome(chromedriver)
        forums_url = f"{game_url}/forums/0"
        driver.get(forums_url)
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        game_data['forums'] = re.sub("\D", "", soup.find('span', class_="nav-pills-count ng-binding ng-scope").text)
        driver.close()
        
        driver = webdriver.Chrome(chromedriver)
        stats_url = f"{game_url}/stats"
        driver.get(stats_url)
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        game_data['num_ratings'] = re.sub("\D", "", soup.find('a', href=ratings_href).text)
        game_data['plays'] = re.sub("\D", "", soup.find('a', href=plays_href).text)
        game_data['page_views'] = re.sub("\D", "", soup.find_all('div', class_="outline-item-description ng-binding")[1].text)
        game_data['owners'] = re.sub("\D", "", soup.find('a', href=owners_href).text)
        driver.close()
        
        data_dict[game_id] = game_data
        games_added += 1
    
        
        
        print(games_added, ' games added')
    
    return data_dict
```

```python
BGG_scraper(BG_urls)
```
